```python
#matlab style
x = np.arange(5)
y = np.random.rand(5)
plt.bar(x,y)
plt.show()
```

```python
#object style
fig,ax = plt.subplots()
ax.bar(x,y)
ax.set_title("Bar Graph")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
#matlab style
x = np.arange(10)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.title("linear")

plt.subplot(2,2,2)
plt.plot(x,x*x)
plt.title("Quadratic")

plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x))
plt.title("Square root")

plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("Log")

plt.tight_layout()
plt.show()
```

```python
#object style
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,x)
ax[0,0].set_title("Linear")
ax[0,1].plot(x,x*x)
ax[0,1].set_title("Quadratic")
ax[1,0].plot(x,np.sqrt(x))
ax[1,0].set_title("Square root")
ax[1,1].plot(x,np.log(x))
ax[1,1].set_title("Log")

plt.subplots_adjust(left=.1,bottom=.1,right=.9,top=.9,wspace=.4,hspace=.4)
plt.show()
```
