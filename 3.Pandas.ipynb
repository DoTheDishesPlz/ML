```python
# Concatinating Joining Merging
df1 = pd.DataFrame([100,200,300,400],index=['a','b','c','d'],columns=['A'])
df1
```

```python
df2 = pd.DataFrame([200,150,50],index=['f','b','d'],columns=['B'])
df2
```

```python
pd.concat((df1,df2),sort=False)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
# Merging 
c = pd.DataFrame([250,150,50],index=["b","c","d"])
df1["C"] = c
df2["C"] = c
print(df1)
print(df2)
```

```python
df1.merge(df2,how='outer',on="C")
```
