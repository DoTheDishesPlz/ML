```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

s1 = pd.Series([1,2,3,4,5])
type(s1)
s2 = pd.Series(np.arange(5,dtype=np.uint8))
s2

arr1 = np.arange(5,dtype=np.uint8)
s3 = pd.Series(arr1,dtype=np.uint16)
#returns values with no index
print(s3.values)
#returns as array
print(s3.array)
print(s3.index)
print(s3.dtype)
print(s3.shape)
print(s3.size)
print(s3.nbytes)
print(s3.ndim)
```
