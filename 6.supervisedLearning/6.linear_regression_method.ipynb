# Linear Regression
#### is a supervised learning model specifically to model the relationship between a dependant variable and one or more independant variable

```python
import numpy as np
import pandas as pd 
from sklearn .linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,mean_squared_error,mean_absolute_error,r2_score
import math

data = pd.DataFrame({
    "marks":[34,51,64,88,95,99],
    "salary":[3400,2900,4250,5000,5100,5600]
})

# returns data in column like array
X = data[["marks"]].values
#returns data back in array like 
y = data["salary"].values
```

```python
X
```

```python
reg = LinearRegression()
reg.fit(X,y)
reg.predict([[70]])
```

```python
reg.predict([[100],[50],[80]])
print(reg.coef_,reg.intercept_)
```

# Visualization

```python
fig,ax = plt.subplots()
fig.set_size_inches(15,7)
plt.scatter(X,y)
ax.axline((0, reg.intercept_), slope=reg.coef_[0], label='regression_label')
ax.legend()
ax.set_xlim(0,110)
ax.set_ylim(1000,10000)

for point in zip(X,y):
    print(point)
    ax.text(point[0][0],point[1]- 500,f"{str(point[0][0]),str(point[1])}")

print(reg.intercept_)
print(reg.coef_)

plt.show()
```

# Evaluating Linear Regression

```python
results_table = pd.DataFrame(data=X,columns=['marks'])
results_table['Predicted Salary'] = reg.predict(X)
results_table["Acutal Salary"] = y
results_table["Error"] = results_table["Acutal Salary"] - results_table["Predicted Salary"]
results_table["Error^2"] = results_table["Error"] **2
```

```python
results_table
```

```python



mean_absolute_error1 = np.abs(results_table["Error"]).mean()
mean_squared_error1 = results_table["Error^2"].mean()
root_mean_squared_mean1 = math.sqrt(mean_squared_error1)
[print(i) for i in ( mean_squared_error1,mean_absolute_error1, root_mean_squared_mean1)]
print("================")
print(mean_squared_error(results_table[["Acutal Salary"]],results_table[["Predicted Salary"]]))
print(mean_absolute_error(results_table["Acutal Salary"],results_table["Predicted Salary"]))
print(math.sqrt(mean_squared_error(results_table[["Acutal Salary"]],results_table[["Predicted Salary"]])))
```

```python
print("R Squared: %.2f" % r2_score(y,reg.predict(X)))
```
