# One Dimensional Data Sets

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import  Polygon
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

mpl.rcParams['font.family'] = 'serif'
%matplotlib inline

np.random.seed(1000)

y = np.random.standard_normal(20)

x = np.arange(len(y))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(),color='b',marker='o',markerfacecolor='r',lw=1.5)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
plt.grid(False)
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

# Two Dimensional Data Sets

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize=(10,6))
plt.plot(y,lw=1.5,color='b',marker='o',markerfacecolor='r')
plt.xlabel('index')
plt.ylabel('value')
```

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize=(10,6))
plt.plot(y[:,0],lw=1.5,color='b',marker='o',markerfacecolor='r',label='1st')
plt.plot(y[:,1],lw=1.5,color='g',marker='o',markerfacecolor='r',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
y[:,0] = y[:,0] * 100
plt.figure(figsize=(10,6))
plt.plot(y[:,0],lw=1.5,color='b',marker='o',markerfacecolor='r',label='1st')
plt.plot(y[:,1],lw=1.5,color='g',marker='o',markerfacecolor='r',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
fig,ax1 = plt.subplots()
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('Simple plot')
ax2 = ax1.twinx()
plt.plot(y[:,1],color='g',marker='o',markerfacecolor='g',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Simple Plot')
plt.subplot(212)
plt.plot(y[:,1],color='g',marker='o',markerfacecolor='g',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width=.5,color='g',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd Data Set')
```

# Scatter Plot

```python
y = np.random.standard_normal((1000,2))

plt.figure(figsize=(10,6))
plt.scatter(y[:,0], y[:,1], marker='o')
plt.xlabel("1st")
plt.ylabel("2nd")
plt.title("Scatter Plot")
```

```python
c = np.random.randint(0,10,len(y))

plt.figure(figsize=(10,6))
plt.scatter(y[:,0],y[:,1],c = c,cmap='coolwarm',marker='o')
plt.colorbar()
plt.xlabel("1st")
plt.ylabel("2nd")
plt.title("Scatter Plot")
```

# Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(y,label=["1st","2nd"],bins=25)
plt.legend(loc=0)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Histogram")
```

```python
plt.figure(figsize=(10,6))
plt.hist(y,label=["1st","2nd"],color=['b','g'],stacked=True,bins=20,alpha=.5)
plt.legend(loc=0)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Histogram")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.boxplot(y)
plt.setp(ax,xticklabels=['1st','2nd'])
plt.xlabel('data set')
plt.ylabel("value")
plt.title("Boxplot")
```

```python
def func(x):
    return 0.5 * np.exp(x)+1
a, b = 0.5,1.5
x = np.linspace(0,2)
y = func(x)
Ix = np.linspace(a,b)
Iy = func(Ix)
verts=[(a,0)]+list(zip(Ix,Iy)) + [(b,0)]
```

```python
fig,ax = plt.subplots(figsize=(10,6))
plt.plot(x,y,'b',linewidth=2)
plt.ylim(bottom=0)
poly = Polygon(verts,facecolor='0.7',edgecolor='0.5')
ax.add_patch(poly)
plt.text(.5*(a+b),1,r'$\int_a^b f(x)\mathrm{d}x$',horizontalalignment='center',fontsize=20)
plt.figtext(.9,.075,'$x$')
plt.figtext(.075,.9,'$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$','$b$'))
ax.set_yticks([func(a),func(b)])
ax.set_yticklabels(('$f(a)$','$f(b)$'))
```

```python
strike = np.linspace(50,150,24)
ttm = np.linspace(.5,2.5,24)
strike, ttm = np.meshgrid(strike,ttm)
strike[:2].round(1)

iv = (strike - 100) ** 2 / (100 * strike) / ttm
iv[:5,:3]

```

```python
fig = plt.figure(figsize=(10,6)).add_subplot(projection="3d")
surf = fig.plot_surface(strike,ttm,iv,rstride=2,cstride=2,cmap=plt.cm.coolwarm,linewidth=.5,antialiased=True)
fig.set_xlabel("strike")
fig.set_ylabel("time to maturity")
fig.set_zlabel("implied volitility")
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111,projection='3d')
ax.view_init(30,60)
ax.scatter(strike,ttm,iv,zdir='z',s=25,c='b',marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time to maturity')
ax.set_zlabel("implied volitity")
```
