# One Dimensional Data Sets

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['font.family'] = 'serif'
%matplotlib inline

np.random.seed(1000)

y = np.random.standard_normal(20)

x = np.arange(len(y))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(),color='b',marker='o',markerfacecolor='r',lw=1.5)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
plt.grid(False)
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

# Two Dimensional Data Sets

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize=(10,6))
plt.plot(y,lw=1.5,color='b',marker='o',markerfacecolor='r')
plt.xlabel('index')
plt.ylabel('value')
```

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize=(10,6))
plt.plot(y[:,0],lw=1.5,color='b',marker='o',markerfacecolor='r',label='1st')
plt.plot(y[:,1],lw=1.5,color='g',marker='o',markerfacecolor='r',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
y[:,0] = y[:,0] * 100
plt.figure(figsize=(10,6))
plt.plot(y[:,0],lw=1.5,color='b',marker='o',markerfacecolor='r',label='1st')
plt.plot(y[:,1],lw=1.5,color='g',marker='o',markerfacecolor='r',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
fig,ax1 = plt.subplots()
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('Simple plot')
ax2 = ax1.twinx()
plt.plot(y[:,1],color='g',marker='o',markerfacecolor='g',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Simple Plot')
plt.subplot(212)
plt.plot(y[:,1],color='g',marker='o',markerfacecolor='g',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(y[:,0],color='b',marker='o',markerfacecolor='b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width=.5,color='g',label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd Data Set')
```
